```r
Cor_check <- function(data){
    M <- data.frame(AV=apply(data,1,mean))
    data_M <- cbind(data, M)
    
    data_alpha <- psych::alpha(data)
    print(data_alpha)
    corrplot::corrplot(cor(data_M))

}

Diff_check <- function(data_jp, data_us, min, max){
    AB_jp <- data_jp
        AB_us <- data_us
    print("Descriptive summary (US)")
    print(summary(apply(AB_us[-ncol(AB_us)], 1,mean)))
    print("Descriptive summary (JP)")
    print(summary(apply(AB_jp[-ncol(AB_jp)], 1,mean)))
    print(c("SD (US)", sd(apply(AB_us[-ncol(AB_us)], 1,mean))))
    print(c("SD (JP)", sd(apply(AB_jp[-ncol(AB_jp)], 1,mean))))
    valueUS <- apply(AB_us[-ncol(AB_us)], 1,mean)
    valueJP <- apply(AB_jp[-ncol(AB_jp)], 1,mean)

print(t.test(valueUS, valueJP))
#print(mean(apply(AB_us[-ncol(AB_jp)], 1,mean)) - mean(apply(AB_jp[-ncol(AB_jp)], 1,mean)))
print(cohen.d(apply(AB_us[-ncol(AB_us)], 1,mean), apply(AB_jp[-ncol(AB_jp)], 1,mean)))

temp <- data.frame(apply(AB_jp[-ncol(AB_jp)], 1,mean))
colnames(temp) <- c("value")
temp$cul <- "JP"
    temp$id <- AB_jp$id
temp2 <- data.frame(apply(AB_us[-ncol(AB_us)], 1,mean))
colnames(temp2) <- c("value")
temp2$cul <- "US"
        temp2$id <- AB_us$id

AB <- rbind(temp, temp2)
    
options(repr.plot.width=3, repr.plot.height=4)
g <- ggplot(AB, aes(x=cul, y=value)) + geom_boxplot()  +
    stat_summary(fun=mean, geom="point", shape=20, size=9, color="blue", fill="red")+ 
    scale_y_continuous(breaks=c(min:max))+
    theme(axis.title.x=element_blank(), 
          axis.title.y=element_blank(),
          axis.text=element_text(size=18)
         )
    print(g)
    
    
    return(AB)
}


AN_jp <- data_jp[str_detect(colnames(data_jp), pattern="AN_")]
AN_jp$id <- data_jp$id
AC_jp <- data_jp[str_detect(colnames(data_jp), pattern="AC_")]
AC_jp$id <- data_jp$id
AC_jp$AC_6. <- 8 - AC_jp$AC_6.
AC_jp$AC_9. <- 8 - AC_jp$AC_9.

RATE_jp <- data_jp[str_detect(colnames(data_jp), pattern="RATE_WITCH")]
RATE_jp$id <- data_jp$id

MR_jp <- 8 - data_jp[str_detect(colnames(data_jp), pattern="MR_")]
MR_jp$id <- data_jp$id
SBS_jp <- data_jp[str_detect(colnames(data_jp), pattern="SBS_")]
SBS_jp$id <- data_jp$id
AM_jp <- data_jp[str_detect(colnames(data_jp), pattern="AM_")]
AM_jp$id <- data_jp$id
IRI_jp <- data_jp[str_detect(colnames(data_jp), pattern="IRI_")]
IRI_jp$id <- data_jp$id
IRI_jp$IRI_3.PT. <- 8 - IRI_jp$IRI_3.PT.
IRI_jp$IRI_4.EC. <- 8 - IRI_jp$IRI_4.EC.
IRI_jp$IRI_14.EC. <- 8 - IRI_jp$IRI_14.EC.
IRI_jp$IRI_15.PT. <- 8 - IRI_jp$IRI_15.PT.
IRI_jp$IRI_18.EC. <- 8 - IRI_jp$IRI_18.EC.
IRI_jp <- IRI_jp[c(1,3,5,7,9,10,12, 2,4,6,8,11,13,14,15)]

EXP_jp <- data_jp[str_detect(colnames(data_jp), pattern="EXP")]
EXP_jp$id <- data_jp$id
RHAB_jp <- data_jp[c("RHAB", "FNL1", "FNL2", "id")]


AN_us <- data_us[str_detect(colnames(data_us), pattern="AN_")]
AN_us$id <- data_us$id
AC_us <- data_us[str_detect(colnames(data_us), pattern="AC_")]
AC_us$id <- data_us$id
AC_us$AC_6. <- 8 - AC_us$AC_6.
AC_us$AC_9. <- 8 - AC_us$AC_9.

RATE_us <- data_us[str_detect(colnames(data_us), pattern="RATE_WITCH")]
RATE_us$id <- data_us$id
MR_us <- 8 - data_us[str_detect(colnames(data_us), pattern="MR_")]
MR_us$id <- data_us$id
SBS_us <- data_us[str_detect(colnames(data_us), pattern="SBS_")]
SBS_us$id <- data_us$id
AM_us <- data_us[str_detect(colnames(data_us), pattern="AM_")]
AM_us$id <- data_us$id
IRI_us <- data_us[str_detect(colnames(data_us), pattern="IRI_")]
IRI_us$id <- data_us$id
IRI_us$IRI_3.PT. <- 8 - IRI_us$IRI_3.PT.
IRI_us$IRI_4.EC. <- 8 - IRI_us$IRI_4.EC.
IRI_us$IRI_14.EC. <- 8 - IRI_us$IRI_14.EC.
IRI_us$IRI_15.PT. <- 8 - IRI_us$IRI_15.PT.
IRI_us$IRI_18.EC. <- 8 - IRI_us$IRI_18.EC.
IRI_us <- IRI_us[c(1,3,5,7,9,10,12, 2,4,6,8,11,13,14,15)]

EXP_us <- data_us[str_detect(colnames(data_us), pattern="EXP")]
EXP_us$id <- data_us$id
RHAB_us <- data_us[c("RHAB", "FNL1", "FNL2", "id")]
```

# Regression analysis

```r
lrt <- function (obj1, obj2) {
  L0 <- logLik(obj1)
  L1 <- logLik(obj2)
  L01 <- as.vector(- 2 * (L0 - L1))
  df <- attr(L1, "df") - attr(L0, "df")
  list(L01 = L01, df = df,
  "p-value" = pchisq(L01, df, lower.tail = FALSE))
}
```

```r
null_model <- lm(MR ~RHAB + SBS +  AM + IRI_PT + IRI_EC+RATE_WITCH + EXP + Age+EDU + Gender1 + Gender2, data=quant_norm)
summary(null_model)
AIC(null_model)


model_1a <- lm(MR ~RHAB + SBS +  (AM + IRI_PT + IRI_EC+RATE_WITCH + EXP + Age+EDU + Gender1 + Gender2)*cul, data=quant_norm)
summary(model_1a)
AIC(model_1a)


model_1b <- lm(MR ~(RHAB + SBS +  AM + IRI_PT + IRI_EC+RATE_WITCH + EXP + Age+EDU + Gender1 + Gender2)*cul, data=quant_norm)
summary(model_1b)
AIC(model_1b)
```

```r
lrt(model_1a, model_1b)

logLik(null_model)
logLik(model_1a)
logLik(model_1b)
```

```r
compare_models <- function(obs, hypo, cor){
    model_ <- paste(obs, hypo, cor, sep="")
    fit_fix_ <- cfa(model_, joint_data, group = "cul", std.lv=TRUE)
    a <- lavTestLRT(fit_preRegi_FULL_multi, fit_fix_)
    return(a)
}

compare_modelsb <- function(obs, hypo, cor){
    model_ <- paste(obs, hypo, cor, sep="")
    fit_fix_ <- cfa(model_, joint_data, group = "cul", std.lv=TRUE)
    a <- lavTestLRT(fit_preRegi_FULL_multi, fit_fix_)
    return(c(fit_fix_, a))
}
```

```r
RAtoACE <- compare_modelsb(model_obs_part, model_RAtoACE_hypo_part, model_cor_part)
RAtoANI <- compare_modelsb(model_obs_part, model_RAtoANI_hypo_part, model_cor_part)
RAtoMR <- compare_modelsb(model_obs_part, model_RAtoMR_hypo_part, model_cor_part)

SBStoACE <- compare_modelsb(model_obs_part, model_SBStoACE_hypo_part, model_cor_part)
SBStoANI <- compare_modelsb(model_obs_part, model_SBStoANI_hypo_part, model_cor_part)
SBStoMR <- compare_modelsb(model_obs_part, model_SBStoMR_hypo_part, model_cor_part)

ACEtoMR <- compare_modelsb(model_obs_part, model_ACEtoMR_hypo_part, model_cor_part)
ANItoMR <- compare_modelsb(model_obs_part, model_ANItoMR_hypo_part, model_cor_part)
```
